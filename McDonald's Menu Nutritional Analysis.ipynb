```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
df = pd.read_csv("Nutrical Dataset.csv") 
```

```python
null_counts = df.isnull().sum()
print("Null counts in each column:")
print(null_counts)
```

```python
calorie_stats = df['Calories'].describe()
print("Descriptive statistics for calorie counts:")
print(calorie_stats)
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Calories'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Calorie Counts')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
nutrients_stats = df[['Total Fat', 'Protein', 'Carbohydrates']].describe()
print("Descriptive statistics for nutritional content:")
print(nutrients_stats)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Total Fat', data=df, palette='muted')
plt.title('Total Fat Content Across Categories')
plt.xlabel('Category')
plt.ylabel('Total Fat')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Protein', data=df, palette='muted')
plt.title('Total Protein Content Across Categories')
plt.xlabel('Category')
plt.ylabel('Protein')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Category', y='Carbohydrates', data=df, palette='muted')
plt.title('Total Carbohydrates Content Across Categories')
plt.xlabel('Category')
plt.ylabel('Carbohydrates')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(df[['Total Fat', 'Protein', 'Carbohydrates']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Nutrients')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Item'], df['Calories'], color='skyblue')
plt.title('Calorie Distribution of Menu Items')
plt.xlabel('Menu Items')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df['Calories'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Calorie Counts')
plt.xlabel('Calories')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Calories'], color='orange')
plt.title('Box Plot of Calorie Counts')
plt.xlabel('Calories')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(df['Item'], df['Protein'], color='skyblue')
plt.title('Protein Distribution of Menu Items')
plt.xlabel('Menu Items')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(8, 6))
plt.hist(df['Calories'], bins=10, color='lightgreen', edgecolor='black')
plt.title('Distribution of Protein Counts')
plt.xlabel('Protein')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

plt.figure(figsize=(8, 6))
sns.boxplot(x=df['Protein'], color='orange')
plt.title('Box Plot of Protein Counts')
plt.xlabel('Protein')
plt.show()
```

```python
avg_fat_category = df.groupby('Category')['Total Fat'].mean()
avg_protein_category = df.groupby('Category')['Protein'].mean()
avg_carbohydrates_category = df.groupby('Category')['Carbohydrates'].mean()
```

```python
plt.figure(figsize=(10, 4), dpi=100)

category = avg_fat_category.index
Fat = avg_fat_category.values
Protein = avg_protein_category.values
Carbohydrates = avg_carbohydrates_category.values

X_axis = np.arange(len(category))

plt.bar(X_axis - 0.2, Fat, 0.2, label = 'Fat', color = 'blue')
plt.bar(X_axis, Protein, 0.2, label = 'Protein', color = 'green')
plt.bar(X_axis + 0.2, Carbohydrates, 0.2, label = 'Carbohydrates', color = 'orange')

plt.xticks(X_axis, category)
plt.xlabel("Menu Categories")
plt.title("Distribution of Fat, Protein and Carbohydrates in McDonald's Foods")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
highest_calorie_item = df.loc[df['Calories'].idxmax()]
print("Menu item with the highest calorie count:")
print(highest_calorie_item)
```

```python
lowest_calorie_item = df.loc[df['Calories'].idxmin()]
print("\nMenu item with the lowest calorie count:")
print(lowest_calorie_item)
```

```python

nutritional_columns = ['Total Fat', 'Protein', 'Carbohydrates']
df[nutritional_columns] = df[nutritional_columns].apply(pd.to_numeric, errors='coerce')

avg_nutritional_content = df.groupby('Category')[nutritional_columns].mean()

print("Average Nutritional Content of Popular Menu Categories:")
print(avg_nutritional_content)
```

```python
avg_nutritional_content.plot(kind='bar', figsize=(10, 6))
plt.title('Average Nutritional Content of Popular Menu Categories')
plt.xlabel('Category')
plt.ylabel('Average Nutritional Content')
plt.xticks(rotation=45)
plt.legend(title='Nutrient')
plt.show()
```
